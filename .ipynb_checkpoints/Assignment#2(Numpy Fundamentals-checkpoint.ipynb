## Question:2

###  How to stack two arrays vertically?

#### Desired Output::

```python
arr1 = np.arange(10).reshape(2, 5).astype('int32')
arr2 = np.ones((2, 5)).astype('int32')
# np.stack((arr1, arr2), axis=0)
np.vstack((arr1, arr2))
```

## Question:3

### How to stack two arrays horizontally?

#### Desired Output::

```python
np.hstack((arr1, arr2))
```
